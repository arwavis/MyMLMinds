# Dividing the dataset into dependent and independent columns

```python
X= df.drop('LUNG_CANCER',axis=1)
y=df['LUNG_CANCER']
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

# Decision Tree Classifier
## Default Hyperparameters

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
decision_tree_model = DecisionTreeClassifier()
```

```python
decision_tree_model.fit(X_train,y_train)
```

```python
decision_tree_pred = decision_tree_model.predict(X_test)
```

# Display the confusion matrix and classification report for your model

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score,plot_confusion_matrix
```

```python
accuracy_score(y_test,decision_tree_pred)
```

```python
confusion_matrix(y_test,decision_tree_pred)
```

```python
plot_confusion_matrix(decision_tree_model,X_test,y_test)
```

```python
print(classification_report(y_test,decision_tree_pred))
```

```python
decision_tree_model.feature_importances_
```

```python
pd.DataFrame(index=X.columns,data=decision_tree_model.feature_importances_,columns=['Feature Importance'])
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(12,8),dpi=150)
plot_tree(decision_tree_model);
```

```python
plt.figure(figsize=(12,8),dpi=150)
plot_tree(decision_tree_model,filled=True,feature_names=X.columns);
```

# Reporting Model Results
To begin experimenting with hyperparameters, let's create a function that reports back classification results and plots out the tree.

```python
def report_model(model):
    model_preds = decision_tree_model.predict(X_test)
    print(classification_report(y_test,model_preds))
    print('\n')
    plt.figure(figsize=(12,8),dpi=150)
    plot_tree(model,filled=True,feature_names=X.columns);
```

```python
pruned_tree = DecisionTreeClassifier(max_depth=2)
pruned_tree.fit(X_train,y_train)
```

```python
report_model(pruned_tree)
```

# Max Leaf Nodes

```python
pruned_tree = DecisionTreeClassifier(max_leaf_nodes=3)
pruned_tree.fit(X_train,y_train)
```

```python
report_model(pruned_tree)
```

# Criterion

```python
entropy_tree = DecisionTreeClassifier(criterion='entropy')
entropy_tree.fit(X_train,y_train)
```

```python
report_model(entropy_tree)
```
